# Linear Regression

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("homeprices.csv")
```

```python
df
```

```python
plt.figure(figsize=(8,5))
plt.scatter(x='area',y='price',data=df,alpha=0.5, color='r')
plt.title('Current Home Prices')
plt.xlabel('Area')
plt.ylabel('Prices')
plt.savefig('Home price.png')
```

```python
from sklearn import linear_model
```

```python
reg = linear_model.LinearRegression()
```

```python
reg.fit(df[['area']],df.price)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
reg.predict([[5000]])
```

```python
plt.figure(figsize=(8,5))
plt.scatter(x='area',y='price',data=df,alpha=0.5, color='r')
plt.plot(df[['area']],reg.predict(df[['area']]))
plt.title('Current Home Prices')
plt.xlabel('Area')
plt.ylabel('Prices')
```
